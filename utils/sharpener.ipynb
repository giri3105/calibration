```python
import cv2
import numpy as np

# Load your blurry image
image = cv2.imread('/home/giri/calibration/images/frame_0008.png') # Replace with your image file
if image is None:
    print("Error: Could not load image.")
    exit()

# Convert to float for processing
float_image = image.astype(np.float32) / 255.0

# Apply Gaussian blur to create a "blurred version" of the image
# sigmaX controls the amount of blur. Larger values mean more blur (and more sharpening effect).
blurred = cv2.GaussianBlur(float_image, (0, 0), sigmaX=2) # (0,0) makes sigmaX dictate kernel size

# Calculate the unsharp mask (original - blurred)
unsharp_mask = float_image - blurred

# Add the unsharp mask back to the original image
# alpha controls the strength of the sharpening. Higher values mean more sharpening.
sharpened = float_image + unsharp_mask * 1.5 # Experiment with this alpha (e.g., 1.0 to 3.0)

# Clip values to [0, 1] and convert back to 8-bit integer
sharpened = np.clip(sharpened, 0, 1)
sharpened_8bit = (sharpened * 255).astype(np.uint8)

cv2.imwrite('Sharpened (Unsharp Mask).png', sharpened_8bit)
```
